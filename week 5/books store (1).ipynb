```python
class Book:
    def __init__(self, book_id, title, price):
        self.book_id = book_id
        self.title = title
        self.price = price

class BookStore:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' added to the stock.")

    def get_book_by_id(self, book_id):
        for book in self.books:
            if book.book_id == book_id:
                return book
        return None

    def get_book_by_title(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                return book
        return None

    def get_book_by_price(self, price):
        for book in self.books:
            if book.price == price:
                return book
        return None

    def show_all_books(self):
        if not self.books:
            print("No books in the stock.")
        else:
            print("Books in the stock:")
            for book in self.books:
                print(f"ID: {book.book_id}, Title: {book.title}, Price: {book.price}")

    def delete_book(self, identifier, value):
        books_to_delete = []

        if identifier == "id":
            books_to_delete = [book for book in self.books if book.book_id == value]
        elif identifier == "title":
            books_to_delete = [book for book in self.books if book.title.lower() == value.lower()]
        elif identifier == "price":
            books_to_delete = [book for book in self.books if book.price == value]

        if not books_to_delete:
            print("No matching books found.")
        else:
            for book in books_to_delete:
                self.books.remove(book)
                print(f"Book '{book.title}' deleted from the stock.")

# Function to display menu and get user choice
def display_menu():
    print("\nOptions:")
    print("1. Add a new book")
    print("2. Get a book by ID")
    print("3. Get a book by Title")
    print("4. Show all books")
    print("5. Delete a book")
    print("6. Quit")
    return input("Enter your choice (1-6): ")

# Main program
bookstore = BookStore()

while True:
    choice = display_menu()

    if choice == "1":
        book_id = int(input("Enter book ID: "))
        title = input("Enter book title: ")
        price = float(input("Enter book price: "))
        new_book = Book(book_id, title, price)
        bookstore.add_book(new_book)
    elif choice == "2":
        book_id = int(input("Enter book ID to search: "))
        book = bookstore.get_book_by_id(book_id)
        if book:
            print(f"Book found: ID: {book.book_id}, Title: {book.title}, Price: {book.price}")
        else:
            print("No matching book found.")
    elif choice == "3":
        title = input("Enter book title to search: ")
        book = bookstore.get_book_by_title(title)
        if book:
            print(f"Book found: ID: {book.book_id}, Title: {book.title}, Price: {book.price}")
        else:
            print("No matching book found.")
    elif choice == "4":
        bookstore.show_all_books()
    elif choice == "5":
        delete_option = input("Enter delete option (id/title/price): ").lower()
        value = input("Enter book id to delete: ")
        bookstore.delete_book(delete_option, value)
    elif choice == "6":
        print("Exiting the program. Thank you!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 7.")
```
